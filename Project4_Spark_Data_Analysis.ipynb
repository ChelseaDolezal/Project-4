```python
# per Customer, calculate RFM values (Recency,Frequency,Monetary)

df_customer_rfm = df_invoice_details.groupby('CustomerID').agg({
    'Recency': 'min',
    'InvoiceNo': 'nunique',  # Frequency
    'TotalPrice': 'sum'      # Monetary
}).reset_index()

# fix column names          CustomerID,   Recency,   InvoiceNo,   TotalPrice
df_customer_rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']

df_customer_rfm.head()
```

```python
# Function to Display Info about the passed-in Customer Segment

def display_segment_info (segment_number, segment_text, customer_id_array_for_segment):

    df_filtered_customer_rfm = df_customer_rfm[df_customer_rfm['CustomerID'].isin(customer_id_array_for_segment)]
    
    segment_average_recency   = df_filtered_customer_rfm['Recency'].mean()
    segment_average_frequency = df_filtered_customer_rfm['Frequency'].mean()
    segment_average_monetary  = df_filtered_customer_rfm['Monetary'].mean()

    segment_average_monetary_str = f"${segment_average_monetary :,.2f}"

    df_filtered_invoice_details = df_invoice_details[df_invoice_details['CustomerID'].isin(customer_id_array_for_segment)]
    
    segment_customers = df_filtered_invoice_details['CustomerID'].nunique()
    segment_records   = df_filtered_invoice_details.shape[0]
    segment_invoices  = df_filtered_invoice_details['InvoiceNo'].nunique()

    print(f"Customer Segment {segment_number} - {segment_text}:")
    print(f" - Customers:         {segment_customers :>12}  ( { (segment_customers / total_customers) * 100 :5.2f}% )")
    print(f" - Records:           {segment_records   :>12}  ( { (segment_records   / total_records  ) * 100 :5.2f}% )")
    print(f" - Invoices:          {segment_invoices  :>12}  ( { (segment_invoices  / total_invoices ) * 100 :5.2f}% )")
    print(f" - average Recency:   {segment_average_recency :12.2f}")
    print(f" - average Frequency: {segment_average_frequency :12.2f}")
    print(f" - average Monetary:  {segment_average_monetary_str :>12}")
    print()
```

```python
# Total Calculations
total_customers = df_invoice_details['CustomerID'].nunique()
total_records   = df_invoice_details.shape[0]
total_invoices  = df_invoice_details['InvoiceNo'].nunique()
```

```python
# Copy spark dataframes ... keep original as-is , and use copy for the tasks that follow ...
sdf_invoices = sdf_invoice_details
sdf_subgroup1 = sdf_customer_segment_0
sdf_subgroup2 = sdf_customer_segment_1
sdf_subgroup3 = sdf_customer_segment_2
sdf_subgroup4 = sdf_customer_segment_3
```

```python
# Rename current 'Recency' column to be 'Recency_original'
sdf_invoices = sdf_invoices.withColumnRenamed('Recency', 'Recency_original')

# Add new empty columns for 'Cluster', 'Recency', 'Frequency', and 'Monetary'
sdf_invoices = sdf_invoices.withColumn('Cluster',   lit(None).cast('integer'))
sdf_invoices = sdf_invoices.withColumn('Recency',   lit(None).cast('double'))
sdf_invoices = sdf_invoices.withColumn('Frequency', lit(None).cast('double'))
sdf_invoices = sdf_invoices.withColumn('Monetary',  lit(None).cast('double'))
```

```python
# Add Info from each sdf_subgroup# datframe into the new columns within the sdf_invoices dataframe

# Step 2: Perform a left join with 'sdf_subgroup1'
joined_sdf_subgroup1 = sdf_invoices.join(
    sdf_subgroup1.select('CustomerID', 'Cluster', 'Recency', 'Frequency', 'Monetary')
                        .withColumnRenamed('Cluster',   'Cluster_subgroup1')
                        .withColumnRenamed('Recency',   'Recency_subgroup1')
                        .withColumnRenamed('Frequency', 'Frequency_subgroup1')
                        .withColumnRenamed('Monetary',  'Monetary_subgroup1'),
    on='CustomerID',
    how='left'
)

# Step 3: Perform a left join with 'sdf_subgroup2'
joined_sdf_subgroup2 = joined_sdf_subgroup1.join(
    sdf_subgroup2.select('CustomerID', 'Cluster', 'Recency', 'Frequency', 'Monetary')
                        .withColumnRenamed('Cluster',   'Cluster_subgroup2')
                        .withColumnRenamed('Recency',   'Recency_subgroup2')
                        .withColumnRenamed('Frequency', 'Frequency_subgroup2')
                        .withColumnRenamed('Monetary',  'Monetary_subgroup2'),
    on='CustomerID',
    how='left'
)

# Step 4: Perform a left join with 'sdf_subgroup3'
joined_sdf_subgroup3 = joined_sdf_subgroup2.join(
    sdf_subgroup3.select('CustomerID', 'Cluster', 'Recency', 'Frequency', 'Monetary')
                        .withColumnRenamed('Cluster',   'Cluster_subgroup3')
                        .withColumnRenamed('Recency',   'Recency_subgroup3')
                        .withColumnRenamed('Frequency', 'Frequency_subgroup3')
                        .withColumnRenamed('Monetary',  'Monetary_subgroup3'),
    on='CustomerID',
    how='left'
)

# Step 5: Perform a left join with 'sdf_subgroup4'
joined_sdf_subgroup4 = joined_sdf_subgroup3.join(
    sdf_subgroup4.select('CustomerID', 'Cluster', 'Recency', 'Frequency', 'Monetary')
                        .withColumnRenamed('Cluster',   'Cluster_subgroup4')
                        .withColumnRenamed('Recency',   'Recency_subgroup4')
                        .withColumnRenamed('Frequency', 'Frequency_subgroup4')
                        .withColumnRenamed('Monetary',  'Monetary_subgroup4'),
    on='CustomerID',
    how='left'
)

# Step 6: Update the columns with the values from the sdf_subgroups
sdf_invoices_final = joined_sdf_subgroup4.withColumn(
    'Cluster',
    coalesce(col('Cluster_subgroup1'), col('Cluster_subgroup2'), col('Cluster_subgroup3'), col('Cluster_subgroup4'))
).withColumn(
    'Recency',
    coalesce(col('Recency_subgroup1'), col('Recency_subgroup2'), col('Recency_subgroup3'), col('Recency_subgroup4'))
).withColumn(
    'Frequency',
    coalesce(col('Frequency_subgroup1'), col('Frequency_subgroup2'), col('Frequency_subgroup3'), col('Frequency_subgroup4'))
).withColumn(
    'Monetary',
    coalesce(col('Monetary_subgroup1'), col('Monetary_subgroup2'), col('Monetary_subgroup3'), col('Monetary_subgroup4'))
).drop(
    'Cluster_subgroup1',   'Cluster_subgroup2',   'Cluster_subgroup3',   'Cluster_subgroup4',
    'Recency_subgroup1',   'Recency_subgroup2',   'Recency_subgroup3',   'Recency_subgroup4',
    'Frequency_subgroup1', 'Frequency_subgroup2', 'Frequency_subgroup3', 'Frequency_subgroup4',
    'Monetary_subgroup1',  'Monetary_subgroup2',  'Monetary_subgroup3',  'Monetary_subgroup4'
)  # Drop temporary columns
```

```python
# from pyspark.sql.functions import avg
# from pyspark.sql.functions import to_date, date_format

from pyspark.sql.functions import sum
from pyspark.sql.functions import first, desc
from pyspark.sql.functions import row_number
from pyspark.sql.functions import format_number
from pyspark.sql.window import Window
```

```python

###############
# INTERESTING #
###############

# Add a Description to the product_preference DataFrame
product_preference_with_description = sdf_invoices_final.groupBy("Cluster", "StockCode").agg(
    first("Description").alias("Description"),  # Use first() to get a generic Description
    sum("Quantity").alias("Total_Quantity"),
    sum("TotalPrice").alias("Total_Revenue")
)

# Define the window specification for ranking items within each Cluster by Total_Quantity
window_spec = Window.partitionBy("Cluster").orderBy(desc("Total_Quantity"))

# Add a row number to each row within each Cluster partition
ranked_product_preference = product_preference_with_description.withColumn(
    "Rank",
    row_number().over(window_spec)
)

# Filter to keep only the top 10 items per Cluster
top_10_per_cluster = ranked_product_preference.filter(col("Rank") <= 10)


# Show the results
# top_10_per_cluster.show(50)
# top_10_per_cluster.show(50, truncate=False)


# Format the Total_Revenue column to 2 decimal places
formatted_df = top_10_per_cluster.withColumn(
    "Total_Revenue",
    format_number("Total_Revenue", 2)
)

print("##########################################")
print("# Top 10 Items by Quantity , per Cluster #")
print("##########################################\n")

# Show the results with formatted Total_Revenue
formatted_df.show(50, truncate=False)
```
